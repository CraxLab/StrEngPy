```python
import numpy as np
import matplotlib.pyplot as plt

'==== Read PEER data with the header lines ===================='
# Read record file with many columns data
file = open('figures_files/RSN1158_KOCAELI_DZC180.AT2', 'r')
lines = file.readlines() # Datayı listeye attık
file.close()

HeaderNo=4 # skipped header number from top the record file
data=[]
for line in lines[HeaderNo:]:
    row = line.split()
    for i in row:
        data.append(float(i))

        
DT = 0.0050
NPTS= 5437
print(len(data))
t = np.arange(0,NPTS*DT,DT)
print(len(t))
print(t)

# Plot Ground Motion
fig, xy = plt.subplots()
xy.plot(t,data, color="r")
xy.set(xlabel='Time (sec)', ylabel='Acceleration (g)', title='t - ag curve')
xy.grid()
plt.show()
```

```python
'==== Read PEER data with the header lines ===================='
# Read record file with many columns data
file = open('figures_files/RSN1158_KOCAELI_DZC180.AT2', 'r')    # Name of the file
HeaderNo = 4   # Number of Header Lines  not to be read
# Δt = 0.01      # Time interval of the record
SF = 1      # Scale Factor; if unit is in g, multiply with 9.81 m/s2
lines = file.readlines() 
file.close() 

deprem = str(lines)

dtindex = deprem.find("DT")  # Find the initial index of "DT":
print(dtindex)
dt = float(deprem[dtindex+5 : dtindex+11]) # Find the index range of value of "DT"
print(dt)

nptsindex = deprem.find("NPTS") # Find the initial index of "NPTS":
print(nptsindex)
npts = int(deprem[nptsindex+7 : nptsindex+12]) #Find the index range of value of "NPTS":
print(npts) 


t = np.array([np.arange(0, (npts*dt)+dt, dt)]).T   # time interval
# print(len(t))
# print(t)


data = [0]
for line in lines[HeaderNo:]:
    row = line.split()          # split the rows
    for i in row:               # loop in the each rows
        data.append(float(i))

acc = np.transpose(np.array([data]))*SF    

# Plot Ground Motion
fig, xy = plt.subplots()
xy.plot(t,acc, color="r")
xy.set(xlabel='Time (sec)', ylabel='Acceleration (m/s2)', title='t - ag curve')
xy.grid()
plt.show()
```
